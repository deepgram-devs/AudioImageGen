```python
import openai
import deepgram
import os
import json
import requests
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
openai.organization = os.getenv('OPENAI_ORGANIZATION')
```

```python
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
deepgram_access_code = os.getenv("DEEPGRAM_ACCESS_CODE")
```

```python
def prompt_image_generator(prompt):
    response = openai.Image.create(
      prompt=prompt,
      n=1,
      size="256x256"
    )
    image_url = response['data'][0]['url']
    return image_url
```

```python
prompt_image_generator('a boy playing in the park artist style')
```

```python
url = "https://api.beta.deepgram.com/v1/listen"
params = {'summarize':'v2'}
payload = {"url": "https://audio.jukehost.co.uk/SY3der7O7rHGAMvGWblqdvP33TPAS64Y.mp3"}
headers = {
    "accept": "application/json",
    "content-type": "application/json",
    "Authorization": f"Token {str(deepgram_access_code)}"
}

response = requests.post(url, json=payload, headers=headers, params= params)
print(json.dumps(response.json()))
```
